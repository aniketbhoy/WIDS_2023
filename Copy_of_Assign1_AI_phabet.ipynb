 **For the DataFrame data, replace all occurrences of the value 'pig' in the 'animal' column with 'boar', but only for rows where 'ounces' is less than 8.**

```python
import pandas as pd

data = pd.DataFrame({
    'animal': ['cat', 'dog', 'pig', 'cow', 'pig', 'cat' , 'cow' , 'cow'],
    'ounces': [5, 12, 8, 7, 6 , 9 , 11 , 3],
    'sound': ['meow', 'woof', 'oink', 'moo' , 'oink', 'meow', 'moo' , 'moo']
})
```

```python
data.loc[data['ounces']<8,'animal'] = data.loc[data['ounces']<8,'animal'].replace('pig','boar')
data
```

 **Select all rows from the DataFrame data where 'animal' is 'cow' and 'ounces' is greater than the mean value of 'ounces'.**


```python
rows = data.loc[(data['animal']=='cow') & (data['ounces']>data['ounces'].mean())]
rows
```

 **Create a new column in the DataFrame data named 'scaled_ounces' such that it contains the values from the 'ounces' column scaled by subtracting the mean and dividing by the standard deviation.**

```python
mean = data['ounces'].mean()
sd = data['ounces'].std()
data['scaled_ounces'] = (data['ounces']-mean)/sd
data
```

**Plot a sine and cosine curve in the same plot. Customize the plot with labels, title, and legend.**

```python
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize = (8,6))
plt.plot(x, y, color = 'red', label='Sine curve')
plt.plot(x, z, color = 'orange', label='Cosine curve')

plt.xlabel('sin(x)')
plt.ylabel('cos(x)')
plt.title('sine and cosine values')
plt.legend()
plt.show()
```

**Simulate a Bernoulli distribution with 100 trials and plot a bar chart of the outcomes.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
np.random.seed(42)
p = 0.6
distribution = bernoulli(p)
outcomes = distribution.rvs(size=100)
x = [0,1]

f = np.sum(outcomes == 0)
s = np.sum(outcomes == 1)

plt.bar(x, [f, s], tick_label=['Outcome=0', 'Outcome=1'])
plt.title('Bernoulli Distribution Simulation')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.show()
```

**Create a normal distribution with mean 0 and standard deviation 1. Plot a histogram and overlay the probability density function (PDF) on top.**


```python
from scipy.stats import norm
mu = 0
sigma = 1
data = np.random.normal(mu,sigma,1000)
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue', label='Histogram')
xmin, xmax = plt.xlim()
x = np.linspace(xmin,xmax,100)
p = norm.pdf(x, mu, sigma)
plt.plot(x, p, 'k', linewidth=2, label='PDF')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
